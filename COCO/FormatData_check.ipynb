```python
import glob
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

from COCOWrapper import COCOWrapper
```

```python
root='/home/gregory/Datasets/COCO'
year='2017'
mode='train'

coco = COCOWrapper(root = root, mode = mode, year = year)

imgs = coco.coco.loadImgs(coco.coco.getImgIds())
```

```python
file_id = imgs[np.random.randint(len(imgs))]['file_name'].split('.')[0]
print(file_id)

files = []
for file in glob.glob('{}/*/{}*'.format(root, file_id), recursive = True):
    files.append(file)

files.sort(reverse = True)

for file in files:
    print(file)
    im = Image.open(file).convert('RGB')
    plt.imshow(im)
    plt.show()
    plt.close()
```
